```python
X = np.array([[2, -1, 0],
              [-1, 2, -1],
              [0, -1, 2]])
y = np.array([1, 1, 1])
for x in np.linalg.solve(X, y):
    print("%lf" % x, end=' ')
```

```python
n = 20.0
m = 8.0
A = []
y = []
for i in range(20):
    A.append([])
    for j in range(20):
        if (i == j):
            A[i].append(n + m ** 2 + float(j) /m + float(i)/n)
        else:
            A[i].append((float(i) + float(j))/(n + m))
    y.append(200.0 + 5.0 * i)

X = np.linalg.solve(A, y)
X
```

```python
for i in X:
    print(i)
```

```python
np.linalg.det(A)
```
